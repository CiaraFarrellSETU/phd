```python
import geopandas as gpd
```

```python
shapefile_path = "/content/Pollardstown_clappied.shp"
gdf = gpd.read_file(shapefile_path)
```

```python
print("Columns:", gdf.columns)
```

```python
if 'id' in gdf.columns:
    unique_ids = gdf['id'].nunique()
    print(f"Number of unique IDs: {unique_ids}")
else:
    # If no explicit ID column, count total features
    total_features = len(gdf)
    print(f"Number of features (rows): {total_features}")
```

```python
unique_ids = gdf['Id'].nunique()
print(f"Number of unique IDs: {unique_ids}")
```

```python
print("Unique IDs:", gdf['Id'].unique())
```

```python
for col in gdf.columns:
    unique_count = gdf[col].nunique()
    print(f"Column '{col}' has {unique_count} unique values")
```

```python
freq_table = gdf['PRIMARY_FO'].value_counts()

# Print the frequency table
print(freq_table)

# If you want it as a DataFrame for nicer formatting:
freq_df = freq_table.reset_index()
freq_df.columns = ['Habitat Type', 'Count']
print
```
