```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load directly from your browser session storage
df = pd.read_csv('/content/habitat_pixels_for_boxplot (1).csv')

# --- 1. Compute strict 5-number summaries per habitat ---
features = ['EVI', 'B2', 'B8']
for f in features:
    print(f"\n==== 5-NUMBER SUMMARY FOR {f} ====")
    summary = df.groupby('HCH_MAPPED')[f].describe(percentiles=[.25, .5, .75])
    print(summary[['min', '25%', '50%', '75%', 'max']])

# --- 2. Build Boxplots ---
sns.set_theme(style="whitegrid")
habitat_order = ['FS1', 'FS2', 'FW2/WL2', 'GS4', 'PF1', 'PF3', 'WD4', 'WN6', 'WN7']

fig, axes = plt.subplots(3, 1, figsize=(11, 15))
colors = ['#14b8a6', '#2563eb', '#dc2626']

for i, feature in enumerate(features):
    sns.boxplot(
        ax=axes[i], x='HCH_MAPPED', y=feature, data=df,
        order=habitat_order, color=colors[i], fliersize=1.5
    )
    axes[i].set_title(f'Habitat Distribution Profile: {feature}', fontsize=12, fontweight='bold')
    axes[i].set_xlabel('Habitat Class Code')
    axes[i].set_ylabel('Spectral/Index Intrinsic Value')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load directly from your browser session storage
df = pd.read_csv('/content/liskeenan_pixels_for_boxplot_growth.csv')

# Drop any missing values in your habitat column to avoid plotting blank categories
df = df.dropna(subset=['HCH_MAPPED'])

# =================================================================
# DYNAMICALLY EXTRACT AND SORT HABITATS
# =================================================================
# This grabs every unique habitat name present in your CSV and sorts them alphabetically
habitat_order = sorted(df['HCH_MAPPED'].unique())

print(f"✅ Dynamically detected {len(habitat_order)} habitats to plot:")
print(habitat_order)

# --- 1. Compute strict 5-number summaries per habitat ---
features = ['EVI', 'B2', 'B8']
for f in features:
    print(f"\n==== 5-NUMBER SUMMARY FOR {f} ====")
    summary = df.groupby('HCH_MAPPED')[f].describe(percentiles=[.25, .5, .75])
    # Reindex using our dynamic order to keep the print tables matching the plot order
    five_num_summary = summary[['min', '25%', '50%', '75%', 'max']].reindex(habitat_order)
    print(five_num_summary)

# --- 2. Build Boxplots ---
sns.set_theme(style="whitegrid")

# Dynamic plot height: scales wider if you have 15 classes so the labels don't bunch up
fig, axes = plt.subplots(3, 1, figsize=(14, 18))
colors = ['#14b8a6', '#2563eb', '#dc2626']

for i, feature in enumerate(features):
    sns.boxplot(
        ax=axes[i],
        x='HCH_MAPPED',
        y=feature,
        data=df,
        order=habitat_order, # Uses the automated dynamic list
        color=colors[i],
        fliersize=1.5
    )
    axes[i].set_title(f'Habitat Distribution Profile: {feature}', fontsize=12, fontweight='bold')
    axes[i].set_xlabel('Habitat Class Code')
    axes[i].set_ylabel('Spectral/Index Intrinsic Value')

    # Slight label tilt so that when you scale up to 15+ classes, the text doesn't overlap
    axes[i].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```
